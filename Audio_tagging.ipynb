```python
def upload_data(dataset_path, hop_length=512, num_segments=5, n_mels=10, n_fft=2048):

  data = {
      'train': {
      'Labels': [],
      'Mel_Spectrogram': []
      },
      'validation': {
      'Labels': [],
      'Mel_Spectrogram': []
      },
      'test': {
      'Labels': [],
      'Mel_Spectrogram': []   
      }
  }

  
  #for folder in sorted(os.listdir(dataset_path)):
  counter = 0
  for file in sorted(os.listdir(dataset_path )):

      signal, sr = librosa.load(dataset_path + '/' + file)
      print(signal.shape[0]/hop_length)

      spectogram = librosa.feature.melspectrogram(signal, sr, n_fft=n_fft, hop_length=hop_length, n_mels=10)
      log_spectrogram = librosa.power_to_db(spectogram)

      if file < 'c':
         #Store the spectrograms
        data['train']['Mel_Spectrogram'].append(log_spectrogram.tolist())
         #store corresponding labels
        data['train']['Labels'].append(new_df.iloc[counter,:].to_json())
      elif file == 'c':
        data['validation']['Mel_Spectrogram'].append(log_spectrogram.tolist())
        data['validation']['Labels'].append(new_df.iloc[counter,:].to_json())
      else:
        data['test']['Mel_Spectrogram'].append(log_spectrogram.tolist())
        data['test']['Labels'].append(new_df.iloc[counter,:].to_json())

      counter += 1
    
    
  #add it to json file 
  with open('/content/d.json', 'w') as f:
    json.dump(data, f, indent=4)
    
    
    #spectrogram = librosa.feature.melspectrogram(y=y, sr=sr) #should the spectrogram be transposed? 
    #Store as JSON file(s)?
    # store the name/id
    #with open(json_path, "w") as fp:
        #json.dump(data, fp, indent=4)
    
```

```python
def conv2d(x, W, b, strides=(2,2)):
    # Conv2D wrapper, with bias and relu activation
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding='SAME')
    x = tf.nn.bias_add(x, b)
    return tf.nn.relu(x)
```

```python
def maxpool2d(x, k=2):
    # MaxPool2D wrapper
    return tf.nn.max_pool(x, ksize=[1, k, k, 1], strides=[1, k, k, 1], padding='valid')
```
